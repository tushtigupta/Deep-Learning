```python
 #Load Pretrained GloVe model

import gensim.downloader as api
import numpy as np
from numpy.linalg import norm
```

```python
#Load 50-dimensional GloVe vectors
glove=api.load("glove-wiki-gigaword-50")
```

```python
##Word Semantic Similarity

similarity1=glove.similarity("king","queen")
similarity2=glove.similarity("king","apple")

print("Similarity (king,queen)",similarity1)
print("Similarity (king,apple)",similarity2)
```

```python
# Sentence Semantic Similarity

#Define Sentence vector Function
def sentence_vector(sentence):
  words=sentence.lower().split()
  vectors=[]

  for word in words:
    if word in glove:
      vectors.append(glove[word])
  return np.mean(vectors, axis=0)
```

```python
#Compare Two sentences
sentence1="I love machine learning"
sentence2="I enjoy arrtificial intelligence"
sentence3="The cat is sleeping"

vec1=sentence_vector(sentence1)
vec2=sentence_vector(sentence2)
vec3=sentence_vector(sentence3)
```

```python
#Cosine similarity function
def cosine_similarity(a,b):
  return np.dot(a,b) /(norm(a)*norm(b))

print("Similarity (s1,s2)",cosine_similarity(vec1,vec2))
print("Similarity (s1,s3)",cosine_similarity(vec1,vec3))
```
